```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])
```

```python
df['Embarked'] = label_encoder.fit_transform(df['Embarked'])
```

```python
df['Pclass'] = label_encoder.fit_transform(df['Pclass'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(df.drop(['Survived'], axis=1), df['Survived'], test_size=0.33)
```

```python
y_test
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier(criterion='entropy', splitter='random')
```

```python
tree.fit(x_train, y_train)
```

```python
prediction = tree.predict(x_test)
prediction
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
con_mat = confusion_matrix(prediction, y_test)
con_mat
```

```python
accuracy_score(prediction, y_test)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log_reg = LogisticRegression(random_state=1)
log_reg.fit(x_train, y_train)
```

```python
prediction = log_reg.predict(x_test)
prediction
```

```python
con_mat = confusion_matrix(prediction, y_test)
con_mat
```

```python
accuracy_score(prediction, y_test)
```
